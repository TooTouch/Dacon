# OPS ( OBP + SLG )
## OBP (출루율)
- https://ko.wikipedia.org/wiki/%EC%B6%9C%EB%A3%A8%EC%9C%A8
- 여기서 데이터에 없는건 희생플라이(SF)
- 공식 : (안타+볼넷+몸에 맞은 공) / (타수+볼넷+몸에 맞은 공+희생플라이)
## SLG (장타율)
- https://ko.wikipedia.org/wiki/%EC%9E%A5%ED%83%80%EC%9C%A8
- 공식 : (단타 수 + 2루타 수 x 2 + 3루타 수 x 3 + 홈런 수 x 4)/타수

```python
years = data[data.team=='KT'].year.unique()
years
```

```python
year_OPS_rating = dict()
teams = ['Hanhwa','Samsung','Lotte','Nexen','Doosan','LG','KIA','SK','KT','NC']
for y in years:
    OPS_by_team = list()
    for t in teams:
        df_t_y = data[(data.year==y) & (data.team==t)].sum()
        OBP = (df_t_y.loc['H'] + df_t_y.loc['BB'] + df_t_y.loc['HBP']) / (df_t_y.loc['AB'] + df_t_y.loc['BB'] + df_t_y.loc['HBP'] + df_t_y.loc['SF'])
        SLG = (df_t_y.loc['H'] + df_t_y.loc['2B'] + 2*df_t_y.loc['3B'] + 3*df_t_y.loc['HR']) / df_t_y.loc['AB']
        OPS = OBP + SLG
        OPS_by_team.append(OPS)
    ts_df = pd.DataFrame({'team':teams,'OPS_AVG':OPS_by_team}).sort_values(by='OPS_AVG', ascending=False)
    year_OPS_rating[y] = list(ts_df.team)
pd.DataFrame(year_OPS_rating)[['2018','2017','2016','2015']]
```

## 자기 상관관계

- 위에서 선택한 ['H', 'G', 'HBP', 'GDP', '2B', 'HR', 'SO', 'R', 'RBI', 'AB', 'TB', 'BB', 'SF', 'MH', 'PA'] 중 autocorrelation이 높은 변수 (기준은 > 0.5)
- **['H', 'HBP', '2B', 'HR', 'SO', 'R', 'RBI', 'AB', 'TB', 'BB', 'MH', 'PA']**

```python
def get_self_corr(var, data, ax):
    x=[]
    y=[]
    data=data.loc[data['AB']>=100,]
    for name in data['batter_name'].unique():
        a=data.loc[data['batter_name']==name,]
        k=[]
        for i in a['year'].unique():
            if (a['year']==i+1).sum()==1:
                k.append(i)
        for i in k:
            x.append(a.loc[a['year']==i,var].iloc[0])
            y.append(a.loc[a['year']==i+1,var].iloc[0])
    ax.scatter(x,y)
    ax.set_title('{0:} ({1:.4f})'.format(var,pd.Series(x).corr(pd.Series(y))))
```

```python
data.year = data.year.astype(int)
data['1B']=data['H']-data['2B']-data['3B']-data['HR']
autocor_features = ['H', 'G', 'HBP', 'GDP', '2B', '3B', 'SB', 'HR', 'SO', 'R', 'CS', 'RBI', 'AB', 'TB', 'BB', 'SF', 'MH', 'PA', 'SAC', 'IBB']
f, ax = plt.subplots(2,10, figsize=(40,7))
for i in range(len(autocor_features)):
    get_self_corr(autocor_features[i], data, ax[i//10,i%10])
```
