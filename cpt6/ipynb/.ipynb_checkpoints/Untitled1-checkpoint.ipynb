```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

import sys
sys.path.append('../code/')

from preprocessing import *
from utils import *
```

```python
# data
regular_day = pd.read_csv('../dataset/Regular_Season_Batter_Day_by_Day.csv', index_col=0)
print('regular_day.shape: ',regular_day.shape)
```

```python
regular_day.info()
```

```python
regular_day2018 = regular_day[regular_day.year==2018]
```

```python
np.sort(regular_day2018.date.unique())
```

```python
regular_day2018.loc[:,'month'] = regular_day2018.date.apply(lambda x: str(x).split('.')[0])
```

```python
names2018 = regular_day2018.batter_name.unique()
avg2018 = pd.DataFrame({'batter_name':names2018})
```

```python
mon3 = regular_day2018[(regular_day2018.month=='3')&(regular_day2018.date==3.31)]
avg2018.loc[avg2018.batter_name.isin(mon3.batter_name.unique()), 'mon3_avg'] = mon3.avg2
mon4 = regular_day2018[(regular_day2018.month=='4')&(regular_day2018.date==4.29)]
avg2018.loc[avg2018.batter_name.isin(mon4.batter_name.unique()), 'mon4_avg'] = mon4.avg2
mon5 = regular_day2018[(regular_day2018.month=='5')&(regular_day2018.date==5.31)]
avg2018.loc[avg2018.batter_name.isin(mon5.batter_name.unique()), 'mon5_avg'] = mon5.avg2
mon6 = regular_day2018[(regular_day2018.month=='6')&(regular_day2018.date==6.30)]
avg2018.loc[avg2018.batter_name.isin(mon6.batter_name.unique()), 'mon6_avg'] = mon6.avg2
mon7 = regular_day2018[(regular_day2018.month=='7')&(regular_day2018.date==7.31)]
avg2018.loc[avg2018.batter_name.isin(mon7.batter_name.unique()), 'mon7_avg'] = mon7.avg2
mon8 = regular_day2018[(regular_day2018.month=='8')&(regular_day2018.date==8.16)]
avg2018.loc[avg2018.batter_name.isin(mon8.batter_name.unique()), 'mon8_avg'] = mon8.avg2
mon9 = regular_day2018[(regular_day2018.month=='9')&(regular_day2018.date==9.29)]
avg2018.loc[avg2018.batter_name.isin(mon9.batter_name.unique()), 'mon9_avg'] = mon9.avg2
mon10 = regular_day2018[(regular_day2018.month=='10')&(regular_day2018.date==10.14)]
avg2018.loc[avg2018.batter_name.isin(mon10.batter_name.unique()), 'mon10_avg'] = mon10.avg2
avg2018 = avg2018.fillna(0)
```

```python
plt.figure(figsize=(50,10))
sns.lineplot(x=np.arange(avg2018.shape[0]), y=avg2018.mon3_avg)
sns.lineplot(x=np.arange(avg2018.shape[0]), y=avg2018.mon4_avg)
```
