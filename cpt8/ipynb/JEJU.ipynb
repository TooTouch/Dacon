# 전체 기준

```python
jeju.head()
```

```python
bar_features = ['sex','age','avg_credit_rat']
job_features = ['job_majorc','job_smallc','job_public','job_profession','job_self','job_none','job_other']
income_features = ['avg_income','med_income']
spend_features = ['avg_spend','avg_foreign_spend']
debt_features = ['avg_debt','avg_debt_credit','avg_debt_noneb','avg_debt_mortgage','avg_debt_deposit','avg_debt_collateral']
money_features = income_features + spend_features + debt_features
```

```python
f, ax = plt.subplots(1,3,figsize=(30,5))
for i in range(len(bar_features)):
    sns.countplot(jeju[bar_features[i]], ax=ax[i])
    ax[i].set_title(bar_features[i])
```

```python
# 원 단위를 만원 단위로 변환
jeju[money_features] = jeju[money_features] / 10000
```

```python
# boxplot features 
f, ax = plt.subplots(2,5, figsize=(20,10))
for i in range(2*5):
    sns.distplot(jeju[money_features[i]], ax=ax[i//5,i%5])
    ax[i//5,i%5].set_title(money_features[i])
```

## Job

```python
f, ax = plt.subplots(1,6,figsize=(40,5))
for i in range(len(job_features)):
    job_bin = jeju[job_features[i]].apply(lambda x: 1 if x > 0 else 0)
    sns.countplot(job_bin, ax=ax[i])
    ax[i].set_title(job_features[i])
```

# 부채가 아예 없다?

```python
jeju.loc[:,'debt_sum'] = jeju[debt_features].sum(axis=1)
```

```python
no_debt = jeju[jeju['debt_sum']==0]
```

```python
no_debt[['zip_cd','sex','age','debt_sum','medium_resid_rat', 'large_resid_rat']].shape
```
